```python
import pandas as pd

df = pd.read_csv("Dummy_Sales_Data_v1.csv")
print(df)

```

```python
# head give first five row of data.If we want 10 rows of data then give 10(or any u want) in an argument.

df.head(10)
```

```python
# tail give info about last five row of data. If we want last 7 rows of data
# then give 7 (or any u want) in an argument of tail().

df.tail(7)
```

```python
#This function is used to get a randomly selected row, column, or both from a dataset.
#And this function can be run without using any argument

df.sample()
```

```python
#This function returns a quick summary of the dataframe. And give us info about how many columns,
# entries, its datatype, memory usage in all data

df.info()
```

```python
#This function returns descriptive statistics about the data.  It gives minimum,maximum, mean, standard deviation
#of the values in numerical columns, and the count of all non-null values in the data.

df.describe()
```

```python
# It gives all the rows df where the Quantity is more than 133. You simply need to pass this condition which u want.

df.query("Quantity>33")
```

```python
#This function returns the list of unique values in a column or series. Instead of applying on
#complete DataFrame, it works only on the selected single column.

df["Status"].unique()
```
